# Report Plots Generator

This notebook runs the unified plotting script and saves all report figures into one folder:
`runs/<RUN_NAME>/report_plots/`.

```python
from pathlib import Path

# Set your run name used in the pipeline outputs
RUN_NAME = "report_run"
DATA_TYPE = "Cleaned"  # Cleaned or Raw

plot_script = Path("src/evaluation/generate_report_plots.py")
assert plot_script.exists(), f"Missing script: {plot_script}"
print("Using script:", plot_script)
```

```python
# Run the generator
%run src/evaluation/generate_report_plots.py --run-name $RUN_NAME --data-type $DATA_TYPE
```

```python
# List generated files
out_dir = Path("runs") / RUN_NAME / "report_plots"
print("Output folder:", out_dir.resolve())
if out_dir.exists():
    files = sorted(out_dir.glob("*.png"))
    for f in files:
        print("-", f.name)
else:
    print("Output folder does not exist yet. Run the previous cell first.")
```

```python
# Optional: preview a few generated figures
from IPython.display import Image, display

preview_names = [
    "04_average_cir_waveform_los_vs_nlos.png",
    "05_model_comparison_bar_chart.png",
    "09_validation_curve.png",
    "10_threshold_optimization_curve.png",
]

for name in preview_names:
    p = out_dir / name
    if p.exists():
        print(name)
        display(Image(filename=str(p)))
    else:
        print(f"Missing: {name}")
```
